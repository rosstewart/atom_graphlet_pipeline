```python
import numpy as np
from sklearn.metrics import roc_auc_score
import os
from scipy import stats

def get_pu_empirical_threshold(labels, preds, p_value=0.01):
    """
    Get empirical threshold for PU learning where "negatives" are unlabeled.
    
    Parameters:
    - labels: 1 for positive, 0 (or -1) for unlabeled
    - preds: predicted probabilities
    - p_value: desired false positive rate on unlabeled data
    
    Returns:
    - threshold: score threshold where p_value fraction of unlabeled samples exceed it
    """
    # Convert labels if they're -1/1
    if -1 in labels:
        labels_binary = (labels + 1) / 2
    else:
        labels_binary = labels
    
    # Get unlabeled predictions
    unlabeled_preds = preds[labels_binary == 0]
    
    if len(unlabeled_preds) == 0:
        return None
    
    # For PU learning, we want the threshold where (1-p_value) of unlabeled are below it
    # This assumes most unlabeled are true negatives
    threshold = np.percentile(unlabeled_preds, (1 - p_value) * 100)
    
    return threshold

def analyze_pu_results(results_dir, p_values=[0.01, 0.05, 0.1]):
    """
    Analyze results for PU learning setting.
    """
    descriptions = {
        'ADP': 'ADP-binding',
        'Allo': 'Allosteric residue',
        'ATP': 'ATP-binding',
        'CA': 'Calcium-binding',
        'Cat': 'Catalytic residue',
        'CD': 'Cadmium-binding',
        'CO': 'Cobalt-binding',
        'CU': 'Copper-binding',
        'DNA': 'DNA-binding',
        'FAD': 'FAD-binding',
        'FE': 'Iron-binding',
        'FMN': 'FMN-binding',
        'GDP': 'GDP-binding',
        'GTP': 'GTP-binding',
        'HEM': 'Heme-binding',
        'Hotspot': 'Protein-protein interaction hotspot',
        'K': 'Potassium-binding',
        'MG': 'Magnesium-binding',
        'MN': 'Manganese-binding',
        'NA': 'Sodium-binding',
        'NAD': 'NAD-binding',
        'Nglyco': 'N-linked glycosylation site',
        'NI': 'Nickel-binding',
        'Phos': 'Phosphorylation site',
        'PLP': 'PLP-binding',
        'PPI': 'Protein-protein interaction site',
        'RNA': 'RNA-binding',
        'UDP': 'UDP-binding',
        'ZN': 'Zinc-binding',
    }
    
    results = {}
    
    for prefix in descriptions.keys():
        labels_file = f"{results_dir}/{prefix}_7_5_atom_avg_4graphlet_svm_labels.npy"
        preds_file = f"{results_dir}/{prefix}_7_5_atom_avg_4graphlet_svm_preds.npy"
        
        if os.path.exists(labels_file) and os.path.exists(preds_file):
            labels = np.load(labels_file)
            preds = np.load(preds_file)
            
            try:
                # Convert labels if needed
                if -1 in labels:
                    labels_binary = (labels + 1) / 2
                else:
                    labels_binary = labels
                
                # For PU learning, AUC is computed only on labeled data
                # But we can still compute it as a performance metric
                auc = roc_auc_score(labels_binary, preds)
                
                # Get thresholds for different p-values
                thresholds = {}
                for p_val in p_values:
                    threshold = get_pu_empirical_threshold(labels, preds, p_val)
                    thresholds[p_val] = threshold
                
                # Calculate statistics
                unlabeled_preds = preds[labels_binary == 0]
                positive_preds = preds[labels_binary == 1]
                
                # Estimate contamination in unlabeled set
                # (what fraction of unlabeled might be positive based on score distribution)
                contamination_estimates = {}
                for p_val, thresh in thresholds.items():
                    if thresh is not None:
                        frac_above = np.mean(unlabeled_preds > thresh)
                        contamination_estimates[p_val] = frac_above
                
                results[prefix] = {
                    'auc': auc,
                    'thresholds': thresholds,
                    'n_positive': len(positive_preds),
                    'n_unlabeled': len(unlabeled_preds),
                    'unlabeled_mean': np.mean(unlabeled_preds),
                    'unlabeled_std': np.std(unlabeled_preds),
                    'positive_mean': np.mean(positive_preds),
                    'positive_std': np.std(positive_preds),
                    'contamination_estimates': contamination_estimates,
                }
                
            except Exception as e:
                print(f"Error processing {prefix}: {e}")
    
    return results

def plot_pu_distribution(labels, preds, prefix, p_values=[0.01, 0.05, 0.1]):
    """
    Plot score distributions for PU learning visualization.
    """
    import matplotlib.pyplot as plt
    
    if -1 in labels:
        labels_binary = (labels + 1) / 2
    else:
        labels_binary = labels
    
    unlabeled_preds = preds[labels_binary == 0]
    positive_preds = preds[labels_binary == 1]
    
    plt.figure(figsize=(10, 6))
    
    # Plot distributions
    bins = np.linspace(0, 1, 50)
    plt.hist(unlabeled_preds, bins=bins, alpha=0.6, label=f'Unlabeled (n={len(unlabeled_preds)})', 
             density=True, color='gray')
    plt.hist(positive_preds, bins=bins, alpha=0.6, label=f'Positive (n={len(positive_preds)})', 
             density=True, color='red')
    
    # Add threshold lines
    colors = ['darkred', 'orange', 'gold']
    for i, p_val in enumerate(p_values):
        threshold = get_pu_empirical_threshold(labels, preds, p_val)
        if threshold is not None:
            plt.axvline(threshold, color=colors[i], linestyle='--', 
                       label=f'p={p_val} threshold ({threshold:.3f})')
    
    plt.xlabel('Prediction Score')
    plt.ylabel('Density')
    plt.title(f'{prefix} - PU Learning Score Distributions')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python

if __name__ == "__main__":
    results_dir = "/home/rcstewart/jose_2016_lossgain_datasets/atom_graphlet_pipeline/crossval_results"
    
    # Analyze all results
    results = analyze_pu_results(results_dir)
    
    # Print formatted results
    print("PU Learning Results with Empirical Thresholds:")
    print("=" * 80)
    
    for prefix, data in sorted(results.items()):
        desc = descriptions.get(prefix, prefix)
        print(f"\n{prefix}: {desc}")
        print(f"  AUC (on labeled data): {data['auc']:.3f}")
        print(f"  Samples: {data['n_positive']} positive, {data['n_unlabeled']} unlabeled")
        print(f"  Score statistics:")
        print(f"    Unlabeled: mean={data['unlabeled_mean']:.3f}, std={data['unlabeled_std']:.3f}")
        print(f"    Positive: mean={data['positive_mean']:.3f}, std={data['positive_std']:.3f}")
        print("  Thresholds (percentiles of unlabeled):")
        for p_val, threshold in data['thresholds'].items():
            if threshold is not None:
                contamination = data['contamination_estimates'][p_val]
                print(f"    p={p_val}: threshold={threshold:.3f} "
                      f"(~{contamination:.1%} of unlabeled exceed this)")
```
